```python
import pandas as pd
```

```python
daraz = pd.read_csv('daraz.csv')
```

```python
daraz['Sold'] = daraz['Sold'].apply(lambda x: x.replace(' sold', ''))
```

```python
daraz['Sold'] = pd.to_numeric(daraz['Sold'], errors='coerce')
```
