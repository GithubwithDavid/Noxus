```python
import seaborn as sbs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

### Distrubution Plot

```python
sbs.displot(gas.USA, kde=False)
```

### Joint Plot

```python
sbs.jointplot(x='Year', y='USA', data=gas, kind='hex')
sbs.jointplot(x='Year', y='USA', data=gas, kind='reg')
sbs.jointplot(x='Year', y='USA', data=gas, kind='kde')
```

### KDE Plot

```python
sbs.kdeplot(gas.USA)
```

### Pair Plots

```python
sbs.pairplot(gas)
```

### Rug Plot

```python
sbs.rugplot(gas.USA)
```

### Styling

```python
sbs.set_style('dark')
sbs.set_context('poster', font_scale=0.4)
sbs.jointplot(x='Year', y='USA', data=gas, kind='reg')
```

### Bar Plots

```python
sbs.set_theme(rc={'figure.figsize':(20,5)})
sbs.barplot(x='Year', y='USA', data=gas)
```

### Count Plot

```python
sbs.countplot(x='USA', data=gas)
```

### Box Plot

```python
fifa = pd.read_csv('fifa_data.csv')[:1000]
sbs.boxplot(x='Age', y='no.', data=fifa)
```

### Violin Plot

```python
sbs.violinplot(x='Age', y='no.', data=fifa)
```

### Strip plot

```python
sbs.set_theme(rc={'figure.figsize':(20,20)})
sbs.stripplot(x='Age', y='no.', data=fifa, jitter=True, hue='International Reputation')
```

### Swarm Plot

```python
sbs.set_theme(rc={'figure.figsize':(20,5)})
sbs.swarmplot(x='Age', y='no.', data=fifa, color='black')
```
